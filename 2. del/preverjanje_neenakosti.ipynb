```python
from sage.all import *
import matplotlib.pyplot as plt
import networkx as nx

%run funkcije.ipynb # Uvozi funkcije generate_random_sb(), modularni_produkt() in dominacijsko_stevilo()

def povezani_grafi_s_premerom(n): # Generira nakljune grafe, dokler ne najde povezanega grafa s premerom vsaj 4
    while True:
        p = 0.5  # Verjetnost povezave med vozlii
        G = nx.erdos_renyi_graph(n, p)  # Generira nakljuni graf z n vozlii
        if nx.is_connected(G) and nx.diameter(G) >= 4:  # Preveri, ali je povezan in ima dovolj velik premer
            return G

def preverjanje_neenakosti(n, j, k):  # Preveri neenakost
    veljavni_pari = []  # Seznam parov, kjer je (G  H)  4
    neveljavni_pari = []  # Seznam parov, kjer je (G  H) > 4
    i = 1
    while i <= k:
        G = povezani_grafi_s_premerom(n)  # Generira graf G
        H = generate_random_sb(j)  # Generira graf H tipa (SB)
        H = H.networkx_graph()  # Pretvori v NetworkX obliko
        GH = modularni_produkt(G, H)  # Izrauna modularni produkt G  H
        gamma_GH = dominacijsko_stevilo(GH)  # Izrauna dominacijsko tevilo produkta
        if gamma_GH <= 4:
            veljavni_pari.append((G, H, gamma_GH))  # Shrani par, e neenakost dri
        else:
            neveljavni_pari.append((G, H, gamma_GH))  # Shrani par, e neenakost ne dri
        print(i)  # Izpie trenutno iteracijo
        i = i + 1
    
    enakosti = [(G, H) for G, H, gamma in veljavni_pari if gamma == 4]  # Pari, kjer (G  H) = 4
    return veljavni_pari, enakosti, neveljavni_pari

def risi_grafe(pari, naslov):  # Funkcija za risanje grafov
    for i, (G, H) in enumerate(pari):
        fig, axes = plt.subplots(1, 2, figsize=(12, 6))  # Pripravi dve sliki za G in H
        axes[0].set_title(f"Graf G (Par {i+1})")  # Naslov za graf G
        nx.draw(G, ax=axes[0], with_labels=True, node_color='lightblue')  # Narie graf G
        axes[1].set_title(f"Graf H (Par {i+1})")  # Naslov za graf H
        nx.draw(H, ax=axes[1], with_labels=True, node_color='lightgreen')  # Narie graf H
        plt.suptitle(naslov)  # Doda glavni naslov
        plt.show()  # Prikae graf

veljavni_pari, enakosti, neveljavni_pari = preverjanje_neenakosti(15, 15, 100)

# Izpie rezultate
print(f"Pari z (G  H)  4: {len(veljavni_pari)}")
print(f"Pari z (G  H) = 4: {len(enakosti)}")
print(f"Pari z (G  H) > 4: {len(neveljavni_pari)}")

# Narie grafe, e obstajajo pari z (G  H) = 4
if enakosti:
    risi_grafe(enakosti, naslov="Pari z (G  H) = 4")

# Narie grafe, e obstajajo pari z (G  H) > 4
if neveljavni_pari:
    risi_grafe(neveljavni_pari, naslov="Pari z (G  H) > 4")
```
